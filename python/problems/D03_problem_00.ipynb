```python
# 예제 입력 및 출력
print(abs(3+4j), abs(-0.0), abs(0.0), abs(-5))
```

```python
# 아래에 코드를 작성해주세요.
def my_abs(x):
    if type(x) == complex:
        hs=(x.imag)
        ss=(x.real)
        return (hs**2+ss**2)**0.5
    elif x > 0:
        return x
    elif x < 0:
        return -x
    else:
        return x**2
            
        
print(my_abs(3+4j), my_abs(-0.0), my_abs(0.0), my_abs(-5))

    
```

```python
# 결과 확인
print(my_abs(3+4j), my_abs(-0.0), my_abs(0.0), my_abs(-5))
```

```python
def teacher_abs(x):
    if type(x) == complex:
        return (x.imag**2 + x.real**2)**0.5
    else:
        if x == 0:
            return x ** 2
        if x < 0:
            return x * -1
        else:
            return x
```

```python
print(teacher_abs(3+4j), teacher_abs(-0.0), teacher_abs(0.0), teacher_abs(-5))
```

# all(x)

> all은 인자로 받는 iterable(range, list)의 모든 요소가 참이면 True, 거짓이 하나라도 있으면 False를 리턴한다.
>
>이와 같은 my_all(x)을 작성해보세요

```python
# 예제 입력 및 출력
print(all([1, 2, 5, '6']))
print(all([[], 2, 5, '6']))
print(all([]))
```

```python
# 아래에 코드를 작성해주세요.
def my_all(x):
    for a in x:
        if bool(a) == False:
            return False
    
    return True
    
print(my_all([1, 2, 5, '6']))
print(my_all([3, [], 2, 5, '6']))
```

```python
# 결과 확인
print(my_all([1, 2, 5, '6']))
print(my_all([3, [], 2, 5, '6']))
print(my_all([]))
print(my_all([0]))
```

```python
def teacher_all(x):
    for elements in x:
        if not elements:
            #element가 없으면
            return False
    return True
```

```python
print(teacher_all([1, 2, 5, '6']))
print(teacher_all([3, [], 2, 5, '6']))
print(teacher_all([]))
print(teacher_all([0]))
```

* 파이썬에서 모든 변수는 객체의 주소를 가지고 있을 뿐입니다. 

```
num = [1, 2, 3]
```

* 위와 같이 변수를 생성하면 num이라는 객체를 생성하고, 변수에는 객체의 주소가 저장됩니다.


* 변경가능한(mutable) 자료형과 변경불가능한(immutable) 자료형은 서로 다르게 동작합니다.

따라서, 복사를 하고 싶을 때에는 다음과 같이 해야한다.

````python
a = [1, 2, 3]
b = a[ : ]       #이렇게 copy
b[0] = 5
print(a)
print(b)
````

````python
a = [1, 2, 3]
b = list(a)
b[0] = 5
print(a)
print(b)
````

* 하지만, 이렇게 하는 것도 일부 상황에만 서로 다른 얕은 복사(shallow copy)입니다.

````python
#중첩딘 상황

a = [1, 2, [1, 2]]     #안에 리스트는 같은 곳을 바라보고 있어...
b = a[:]
b[2][0] = 3
print(a)
````

* 만일 중첩된 상황에서 복사를 하고 싶다면, 깊은 복사(deep copy)를 해야합니다. 

* 즉, 내부에 있는 모든 객체까지 새롭게 값이 변경됩니다.

````python
import copy
a = [1, 2, [1, 2]]
b = copy.deepcopy(a)
b[2][0] = 3
print(a)
````

# List Comprehension

List를 만들 수 있는 간단한 방법이 있습니다.

`[ 식 for 변수 in 리스트 ]`

`list(식 for 변수 in 리스트)` => # not good

- 리스트 컴프리헨션의 반복은 인터프리터 내부에서 C언어의 속도로 실행되기 때문에,

````python
even_list = []
for x in range(1, 6):
    even_list.append(x*2)
print(even_list)
````

## 연습 문제
### 짝짓기 - 곱집합

> 주어진 두 list의 가능한 모든 조합을 담은 `pair` 리스트를 만들어주세요.

1. 반복문 활용
2. list comprehension 활용

---

```python
girls = ['jane', 'iu', 'mary']
boys = ['justin', 'david', 'kim']

예시 출력)
    
[('justin', 'jane'), ('justin', 'iu'), ('justin', 'mary'), ('david', 'jane'), ('david', 'iu'), ('david', 'mary'), ('kim', 'jane'), ('kim', 'iu'), ('kim', 'mary')]
```



````python
# 아래에 반복문을 활용하여 만들어주세요.
girls = ['jane', 'iu', 'mary']
boys = ['justin', 'david', 'kim']
pair = []
for a in girls:
    for b in boys:
        pair.append((a, b))
            
print(pair)
````

````python
# 아래에 List comprehension을 활용하여 만들어주세요.
girls = ['jane', 'iu', 'mary']
boys = ['justin', 'david', 'kim']
pair = [(a , b) for a in boys
           for b in girls]
print(pair)
````

### 피타고라스 정리

> 주어진 조건(x < y < z < 50) 내에서 피타고라스 방정식의 해를 찾아보세요.

1. 반복문 활용

2. list comprehension 활용

```
예시 출력)
[(3, 4, 5), (5, 12, 13), (6, 8, 10), (7, 24, 25), (8, 15, 17), (9, 12, 15), (9, 40, 41), (10, 24, 26), (12, 16, 20), (12, 35, 37), (15, 20, 25), (15, 36, 39), (16, 30, 34), (18, 24, 30), (20, 21, 29), (21, 28, 35), (24, 32, 40), (27, 36, 45)]
```

````python
# 아래에 반복문을 활용하여 만들어주세요.
ptg = []
for x in range(50):
    for y in range(x+1, 50):
        for z in range(y+1, 50):
            if x**2 + y**2 == z**2:
                ptg.append((x,y,z))
print(ptg)
````

````python
# 아래에 List comprehension을 활용하여 만들어주세요.
ptg = [(x,y,z) for x in range(50)
          for y in range(x+1, 50)
          for z in range(y+1, 50)
          if x**2 + y**2 == z**2]
print(ptg)
````

## SLICE
### 부분 집합에 대한 참조
- `시작:끝:간격` 구조/ `start:end:step`
- 시작 인데스 부터 끝 인덱스까지, 설정한 간격만큼 건너뛴 원소를 선택한다. (끝 인덱스는 포함x)
- `시작`은 끝이 있는 경우 생략할 수 있고 이때는 0이 된다.
- `끝`은 생략할 수 있으며, 생략하는 경우 리시트의 길이이다. (마지막 인덱스).
- `간격`은 생략할 수 있으며, 생략하는 경우 1이다.
- `간격이 음수`이면 만들어지는 결과는 뒤에서부터 추출하여 역순이 된다.

````python
slice_list = list(range(1, 11))
print(slice_list)
````

````python
# 처음부터 3번 인덱스까지
slice_list[0:4]
````

````python
# 처음부터 끝까지 두칸 간격으로
slice_list[::2]
````

````python
# 시작값을 생략해버리면 리스트의 처음부터라는 의미
slice_list[:5]
````

3. Person 클래스를 정의하고,
이름이 ‘홍길동’, 나이가 20인 p1 인스턴스 객체를 만들어보세요.
이름이 ‘둘리’, 나이가 0인 p2 인스턴스 객체를 만들어보세요.

```python
class Person:
    def __init__(self, name, age = 0):
        self.name = name
        self.age = age
    def greeting(self):
        print(f'안녕하세요. {self.name}. {self.age}살입니다.')
```

```python
p1 = Person('홍길동', 20)
p1.greeting()
```

```python
p2 = Person('둘리')
p2.greeting()
```
